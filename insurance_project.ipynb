```python
rpr_or_rplc=pd.get_dummies(df_claims['Repair_or_Replace'],drop_first=True)
```

```python
rpr_or_rplc.head()
```

```python
df_claims.drop(['Repair_or_Replace'],axis=1,inplace=True)
```

```python
df_claims=pd.concat([df_claims,rpr_or_rplc],axis=1)
```

```python
df_claims
```

```python
df_claims.columns=['claim_id','claim_date','claim_amount','Dealer_ID','Cust_ID','Part_ID','Target']
```

```python
df_claims
```

```python
df_claims.drop(['claim_id','Dealer_ID','Cust_ID','claim_date'],axis=1,inplace=True)
```

```python
df_claims
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_claims.drop('Target',axis=1),
                                               df_claims['Target'],test_size=0.20,random_state=60)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
cross_val_score(classifier,df_claims.drop('Target',axis=1),df_claims['Target'],cv=10,scoring='accuracy').mean()
```
